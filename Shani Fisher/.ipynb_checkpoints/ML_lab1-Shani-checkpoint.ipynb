```python
import pandas as pd
import numpy as np
import re
```

```python
order=pd.read_csv('Orders.csv')
```

```python
order.describe
```

```python
order.dtypes
```

```python
order['Profit']=order['Profit'].str.replace('$', '')
order['Profit']=order['Profit'].str.replace(',', '')
```

```python
order['Profit']=pd.to_numeric(order['Profit'])
```

```python
order['Sales']=order['Sales'].str.replace('$', '')
order['Sales']=order['Sales'].str.replace(',', '')
```

```python
order['Sales']=pd.to_numeric(order['Sales'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
order.sample(10)
```

```python
order['Order.Date']=pd.to_datetime(order['Order.Date'])
order['Order.Date']
order['Order_year']=order['Order.Date'].dt.year
order['Order_month']=order['Order.Date'].dt.month
order.sample(10)
```

```python
order.groupby('Order_month')['Order.ID'].count().plot.bar(color='b')
#looks like in September and Novermber and December many people are buying stuff the season of back to school and holiday
```

```python
#import seaborn

order.groupby('Sub.Category')['Order_month'].count().plot.bar(color='b')#wanted to color by month but couldnt figure it out
#fg = seaborn.FacetGrid(data=order, hue='Order_month',  aspect=1.61)
#fg.map(pyplot.scatter, 'Catgory', 'Number').add_legend()
```
